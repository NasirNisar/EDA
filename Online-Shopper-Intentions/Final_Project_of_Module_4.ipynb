   #         ONLINE SHOPPERS INTENTION

##  Data Set Information:

The dataset consists of feature vectors belonging to 12,330 sessions. The dataset was formed so that each session would belong to a different user in a 1-year period to avoid any tendency to a specific campaign, special day, user profile, or period.

## Data Attribute:


The dataset consists of 10 numerical and 8 categorical attributes. The 'Revenue' attribute can be used as the class label.



| Features          | Description                                                                           |
| -------------- | ------------------------------------------------------------------------------------- |
| Administrative    | Number of different pages visited to the administrativve concern of the website                                           |
| Administrative_Duration      | Total amount of time spent by the visitor on account management related pages                                                   |
| Informational          | Number of different pages visited related to the information of the website and other useful content of the website |
| Informational_Duration | Total amount of time spent by the visitor on informational page                                         |
| ProductRelated       | Number of different pages visited related to different product of the website               |
| ProductRelated_Duration            | Total amount of time spent by the visitor in product related pages                                                                      |
| BounceRates      | average bounce rate value of the pages visited by the visitor                                                                |
| ExitRates         | Average exit rate value of the pages visited by the visitor                                                              |
| PageValues     |  Page values is a average value for a page that a user visited before making a transactions      |
| SpecialDay | Special days indicates the closeness of the siye visiting time to specific special day         |
| Month | Month of the year     |
| OperatingSystems | Operating system of the visitor             |
| Browser | ID of browser from which of the session took place      |
| Region | ID of Region from which of the session took place   |
| TrafficType | ID of different types of sources from which the user landed on the website       |
| VisitorType | Types of Visitors       |
| Weekend | Whether the session was on a Weekend or not         |
|Revenue | Whether the user contributed to the revenue by purchasing or not    |

## K-Means Clusters

```python
x = osi.iloc[:, [1, 6]].values

# checking the shape of the dataset
x.shape


from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters = i,
              init = 'k-means++',
              max_iter = 300,
              n_init = 10,
              random_state = 0,
              algorithm = 'elkan',
              tol = 0.001)
    km.fit(x)
    labels = km.labels_
    wcss.append(km.inertia_)
    
plt.rcParams['figure.figsize'] = (10, 5)
plt.plot(range(1, 11), wcss)
plt.grid()
plt.tight_layout()
plt.title('The Elbow Method', fontsize = 20)
plt.xlabel('No. of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
km = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_means = km.fit_predict(x)

plt.scatter(x[y_means == 0, 0], x[y_means == 0, 1], s = 100, c = 'brown', label = 'Un-interested Customers')
plt.scatter(x[y_means == 1, 0], x[y_means == 1, 1], s = 100, c = 'black', label = 'General Customers')
plt.scatter(x[y_means == 2, 0], x[y_means == 2, 1], s = 100, c = 'grey', label = 'Target Customers')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:, 1], s = 50, c = 'blue' , label = 'centeroid')

plt.title('Administrative Duration vs Duration', fontsize = 20)
plt.grid()
plt.xlabel('Administrative Duration')
plt.ylabel('Bounce Rates')
plt.legend()
plt.show()
```

## Data Pre-Processing

```python
# one hot encoding 

data = pd.get_dummies(osi)

data.columns
```

### Label Encoder

```python
# label encoding of revenue

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
osi['Revenue'] = le.fit_transform(osi['Revenue'])
osi['Revenue'].value_counts()
```

```python
# getting dependent and independent variables

x = data
# removing the target column revenue from x
x = x.drop(['Revenue'], axis = 1)

y = osi['Revenue']

# checking the shapes
print("Shape of x:", x.shape)
print("Shape of y:", y.shape)
```

### Train_Test_Splitting

```python
# splitting the data

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)

# checking the shapes

print("Shape of x_train :", x_train.shape)
print("Shape of y_train :", y_train.shape)
print("Shape of x_test :", x_test.shape)
print("Shape of y_test :", y_test.shape)
```

### Random Forest Classification

```python
# MODELLING

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

model = RandomForestClassifier()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

# evaluating the model
print("Training Accuracy :", model.score(x_train, y_train))
print("Testing Accuracy :", model.score(x_test, y_test))

# confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.rcParams['figure.figsize'] = (6, 6)
sns.heatmap(cm ,annot = True)

# classification report
cr = classification_report(y_test, y_pred)
print(cr)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
#lr = LogisticRegression().fit(x_train, y_train) 
#prediction = lr.predict(x_test)

model = LogisticRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

# evaluating the model
print("Training Accuracy :", model.score(x_train, y_train))
print("Testing Accuracy :", model.score(x_test, y_test))

# confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.rcParams['figure.figsize'] = (6, 6)
sns.heatmap(cm ,annot = True)

# classification report
cr = classification_report(y_test, y_pred)
print(cr)
```

### Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
dtc = DecisionTreeClassifier(criterion="entropy", max_depth=3)
dtc.fit(x_train,y_train)
```

```python
plt.figure(figsize=(20,10))
plot_tree(dtc,fontsize=9,filled=True)
plt.show()
```

### Permutation Importance

```python
# finding the Permutation importance

import eli5
from eli5.sklearn import PermutationImportance

perm = PermutationImportance(model, random_state = 0).fit(x_test, y_test)
eli5.show_weights(perm, feature_names = x_test.columns.tolist())
```

### PDP Visualization

```python
# plotting the partial dependence plot for adminisrative duration

# importing pdp
from pdpbox import pdp, info_plots

base_features = x_test.columns.values.tolist()

feat_name = 'Administrative_Duration'
pdp_dist = pdp.pdp_isolate(model=model, dataset=x_test, model_features = base_features, feature = feat_name)

pdp.pdp_plot(pdp_dist, feat_name)
plt.show()
```

```python
# plotting partial dependency plot for Informational Duration

base_features = x_test.columns.tolist()

feat_name = 'Informational_Duration'
pdp_dist = pdp.pdp_isolate(model, x_test, base_features, feat_name)

pdp.pdp_plot(pdp_dist, feat_name)
plt.show()
```
